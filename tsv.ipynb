```python
import os
import csv

# ===== 경로 설정 =====
base_dir = "/workspace/jiwon/Deepfake_killer-Video-image/train_data"            # train_data 폴더
output_tsv = "/workspace/jiwon/Deepfake_killer-Video-image/train_data/metadata.tsv"
# ====================

entries = []

for label_name in ["real", "fake"]:
    target_dir = os.path.join(base_dir, label_name)
    
    if not os.path.exists(target_dir):
        print(f"⚠ 경고: 폴더 없음 → {target_dir}")
        continue
    
    for fname in os.listdir(target_dir):
        if fname.lower().endswith(".mp4"):
            full_path = os.path.join(target_dir, fname)
            entries.append([full_path, label_name])

# ===== TSV 저장 =====
with open(output_tsv, "w", newline="") as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerows(entries)

print(f"✅ TSV 생성 완료: {output_tsv}")
print(f"총 {len(entries)}개 생성됨")
```
